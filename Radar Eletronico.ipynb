```python
#RADAR ELETRONICO:

velocidade = float(input('Qual é a velocidade atual do carro? '))

#SE A VELOCIDADE FOR MAIOR QUE 50KM/H RECEBE UMA MULTA DE R$ 7,00 POR CADA KM ACIMA DO LIMITE.
if velocidade > 50:
  print('MULTADO! Você está acima da limite permitido que é 50km/h!')
  multa=(velocidade-50)*7
  print(f'Você deve pagar uma multa de R$ {multa:.2f}!')  

#SENÃO, EXIBIR NA TELA UMA MENSAGEM.
print('Tenha um bom dia! Dirija com segurança!')
```
